# Matplotlib - Plot Appearance

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y)
```

```python
ax.plot(x, y, color='orange', linewidth=0.5, alpha=1, linestyle='--', marker='+', markersize=10)
fig
```

```python
ax.set_xlim([0,1])
ax.set_ylim([0, 2])
fig
```

```python
plt.scatter(x, y)
```

```python
plt.hist(x)
```

```python
plt.boxplot(x)
```
