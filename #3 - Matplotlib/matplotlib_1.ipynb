# Matplotlib - 1

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
plt.plot(x, y)
```

```python
plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, y, 'r-')
```

```python
plt.plot(x, y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r')

plt.subplot(1, 2, 2)
plt.plot(y, x, 'b')
```

```python
fig = plt.figure()
```

```python
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(x, y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Set Title')
fig
```

```python
fig = plt.figure()
```

```python
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])
fig
```

```python
axes1.plot(x, y)
axes2.plot(y, x)
fig
```

```python
axes1.set_title('Larger Plot')
axes2.set_title('Smaller Plot')
fig
```

```python
fig, axes = plt.subplots(1, 2)
```

```python
fig, axes = plt.subplots(3, 3)
```

```python
fig, axes = plt.subplots(3, 3)
plt.tight_layout()
```

```python
fig, axes = plt.subplots(1, 2)
for ax in axes:
  ax.plot(x, y)
```

```python
fig, axes = plt.subplots(1, 2)
axes[0].plot(x, y)
axes[0].set_title('First Plot')

axes[1].plot(y, x)
axes[1].set_title('Second Plot')
```

```python
fig = plt.figure(figsize=(8, 2), dpi=100)
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y)
```

```python
fig, axes = plt.subplots(2, 1, figsize=(8, 2), dpi=100)
axes[0].plot(x, y)
axes[1].plot(y, x)
```

```python
fig.savefig('my_picture.png')
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, x**2, label = 'X^2')
ax.plot(x, x**3, label = 'X^3')

ax.legend()
```
